```python
import zipfile, os
import numpy as np
import filetype
from filetype.types.document import Docx
```

```python
def load_train_zip(zip_path: str):
    contents, labels, names = [], [], []
    with zipfile.ZipFile(zip_path) as z:
        for name in z.namelist():
            if name.endswith("/") or name.endswith(".labels"):
                continue
            base = os.path.basename(name)
            if not (base.endswith(".0") or base.endswith(".1")):
                continue
            lab = int(base.rsplit(".", 1)[1])
            contents.append(z.read(name))
            labels.append(lab)
            names.append(name)
    return contents, names, np.asarray(labels, dtype=int)
```

```python
def load_test_zip(zip_path: str):
    names, contents = [], []
    with zipfile.ZipFile(zip_path) as z:
        for name in z.namelist():
            if name.endswith("/") or name.endswith(".labels"):
                continue
            contents.append(z.read(name))
            names.append(name)
    return names, contents
```

```python
contents, names, labels = load_train_zip("docx-train.zip")
cont2, name2, lab2 = load_train_zip("pdf-train.zip")
cont3, name3, lab3 = load_train_zip("rtf-train.zip")

contents = contents + cont2 + cont3
names = names + name2 + name3
labels = np.concatenate([labels, lab2, lab3])
```
